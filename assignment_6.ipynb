1. $F(x) = \begin{cases}
0, & x \le 0 \\
\sqrt{x}, & 0 \le x \le 1 \\
1, & x \ge 1 
\end{cases}
$


```python
#pip install scipy
```

```python
x = np.linspace(-.25, 1.25, 200)
F = []
f = []

for xi in x:
    # CDF
    if xi <= 0:
        F.append(0)
    elif xi >= 1:
        F.append(1)
    else:
        F.append(np.sqrt(xi))
    # PDF
    if 0 < xi < 1:
        f.append(1 / (2 * np.sqrt(xi)))
    else:
        f.append(0)

plt.plot(x, F); plt.title("Distribution 1"); plt.show()
plt.plot(x, f); plt.title("Density 1"); plt.show()
```

3. For $ a < b < c$,
$
F(x) = \begin{cases}
0, & x \le 0 \\
\frac{(x-a)^2}{(b-a)(b-c)}, & a \le x \le c \\
1 - \frac{(b-x)^2}{(b-a)(b-c)}, & c < x < b \\
1, & x \ge b
\end{cases}
$


```python
a= 0
c = 0.6
b=  1.5

x = np.linspace(-0.25, 1.75, 200)
F = []
f = []

for xi in x:
    # CDF
    if xi <= a:
        F.append(0)
    elif a < xi <= c:
        F.append((xi - a)**2 / ((b - a) * (c - a)))
    elif c < xi < b:
        F.append(1 - (b - xi)**2 / ((b - a) * (b - c)))
    else:
        F.append(1)
    # PDF
    if a < xi < c:
        f.append(2 * (xi - a) / ((b - a) * (c - a)))
    elif c < xi < b:
        f.append(2 * (b - xi) / ((b - a) * (b - c)))
    else:
        f.append(0)

plt.plot(x, F); plt.title("Distribution 3"); plt.show()
plt.plot(x, f); plt.title("Density 3 "); plt.show()
```

- Logistic distribution (similar to normal): 
$$
F(x; \sigma) = \dfrac{1}{1+e^{-x/\sigma}} 
$$
with $\sigma >0$.




- support :
 (-infinity, infinity) because this distribtution can take any real number


- density distribution
$$
\frac{dF}{dx} = \frac{e^{-x/\sigma}}{\sigma \left(1 + e^{-x/\sigma}\right)^2}
$$

- plot density + distribution

```python
x = np.linspace(-5, 5, 200) #support
sigma = [0.5, 1, 2, 100] #probs dont need 100 lol

fig, axes = plt.subplots(1, 2, figsize=(12, 5))  # 1 row, 2 columns
for s in sigma:
    #CDF
    F = 1/(1 + np.exp(-x/s))
    axes[0].plot(x, F, label=f'={s}')
    #PDF
    f = np.exp(-x/s) / (s * (1 + np.exp(-x/s))**2)
    axes[1].plot(x, f, label=f'={s}' )

axes[0].set_title("Logistic Distribution")
axes[0].set_xlabel("x")
axes[0].set_ylabel("F(x)")
axes[0].legend()

axes[1].set_title("Logistic Density")
axes[1].set_xlabel("x")
axes[1].set_ylabel("f(x)")
axes[1].legend()

plt.show() 
```


- Negative Binomial (similar to Poisson): The probability mass function for positive integers is:
$$
f(k;r,p) = \dfrac{(k+r-1)!}{k!(r-1)!}(1-p)^k p^r, \quad \text{ for $k=0,...n$}
$$
You can interpret this as follows: Flip a coin that comes up heads with probability $p$ until you get $r$ heads, and then stop. What is the probability of stopping at each $k=0,1,2,...$?


- suppot: [0,infinity) but discrete ints

```python


n_samples = 1000
r = 5   
p = 0.5 
x_max = sp.nbinom.ppf(0.999, r, p) 
x_discrete = np.arange(int(x_max) + 1) 
sample = sp.nbinom.rvs(n=r, p=p, size=n_samples)

#CDF vs ECDF 
# True CDF 
cdf = sp.nbinom.cdf(x_discrete, n=r, p=p)

#ECDF calc
sample_sorted = np.sort(sample)
y_ecdf = np.arange(1, n_samples + 1) / n_samples

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

axes[0].step(x_discrete, cdf, where='post', label=f'True CDF')
axes[0].step(sample_sorted, y_ecdf, where='post', label=f'ECDF (n={n_samples})', linestyle='--')
axes[0].set_title(f"CDF vs ECDF for Negative Binomial ($r$={r}, $p$={p})")
axes[0].set_xlabel("Number of Failures (k)")
axes[0].set_ylabel("F(k)")
axes[0].legend()

#PMF vs KDE 
#True PMF (cuz its in steps not contiguous)
pmf = sp.nbinom.pmf(x_discrete, n=r, p=p)

#KDE 
kde = gaussian_kde(sample)
y_kde = kde(x_discrete) 

axes[1].bar(x_discrete, pmf, width=0.4, label=f'True PMF ($r$={r}, $p$={p})')
axes[1].plot(x_discrete, y_kde, label=f'KDE (n={n_samples})', linestyle='--')

axes[1].set_title(f"PMF vs KDE for Negative Binomial ($r$={r}, $p$={p})")
axes[1].set_xlabel("Number of Failures (k)")
axes[1].set_ylabel("Probability / Density")
axes[1].legend()

plt.show()
```


- Categorical (similar to Bernoulli): The probability mass function over $k = 1, 2, ..., K$ categories is
$$
f(k;p_1,...,p_K) = p_1^{k=1}p_2^{k=2}...p_K^{k=K}
$$
where $0 \le p_i \le 1$ and $ \sum_{k=1}^K p_k = 1$.


- support: [1,k] discrete ints

```python

#  Distribution Parameters 
n_samples = 1000
probabilities = np.array([0.1, 0.2, 0.3, 0.4])
K = len(probabilities) 
category_labels = [f'k= {i+1}' for i in range(K)]
x_discrete = np.arange(1, K + 1) 

sample = sp.rv_discrete(values=(x_discrete, probabilities)).rvs(size=n_samples)

#  CDF vs ECDF 
# True CDF
cdf = np.cumsum(probabilities)

# ECDF 
counts = np.array([np.sum(sample == k) for k in x_discrete])
ecdf_pmf = counts / n_samples
ecdf = np.cumsum(ecdf_pmf)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# CDF 
axes[0].step(x_discrete, cdf, where='post', label='True CDF')
axes[0].step(x_discrete, ecdf, where='post', label=f'ECDF (n={n_samples})', linestyle='--')
axes[0].set_title(f"CDF vs ECDF for Categorical Distribution (K={K})")
axes[0].set_xlabel("Category")
axes[0].set_ylabel("F(k)")
axes[0].legend()
axes[0].grid(True, alpha=0.5)
axes[0].set_xticks(x_discrete)
axes[0].set_xticklabels(category_labels)

# --- PMF vs EPMF (Replacing KDE with EPMF) ---
# True PMF
pmf = probabilities

# EPMF
epmf = ecdf_pmf

width = 0.35
axes[1].bar(x_discrete - width/2, pmf, width, label='True PMF', align='center', color='blue', alpha=0.6)
axes[1].bar(x_discrete + width/2, epmf, width, label=f'EPMF (n={n_samples})')

axes[1].set_title(f"PMF vs EPMF for Categorical Distribution (K={K})")
axes[1].set_xlabel("Category")
axes[1].set_ylabel("Probability")
axes[1].legend()
axes[1].grid(True, alpha=0.5)
axes[1].set_xticks(x_discrete)
axes[1].set_xticklabels(category_labels)

plt.plot()
```
